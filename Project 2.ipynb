The following cell is to filter the dataframe to only find students that have an interest level of '5.0'.

## **Insight #1:**

This cell is used to create a graph of what percentage of students answered 'yes' to each category pertaining to how they recieved information about CCM computing courses.

```python
#list for percentage of "Yes"'s from each column pertaining to how students received info about CCM computing courses'
df3 = [(len(df[df["info_from_counselor"] == "Yes"]))/len(df)*100,
      (len(df[df["info_from_teacher"] == "Yes"])/len(df)*100),
      (len(df[df["ccm_it_website"] == "Yes"])/len(df)*100),
      (len(df[df["info_from_admissions"] == "Yes"])/len(df)*100),
      (len(df[df["info_from_advisor"] == "Yes"])/len(df)*100),
      (len(df[df["info_from_employer"] == "Yes"])/len(df)*100),
      (len(df[df["ccm_workforce_dev"] == "Yes"])/len(df)*100),
      (len(df[df["nj_workforce_dev"] == "Yes"])/len(df)*100),
      (len(df[df["info_from_misc"] == "Yes"])/len(df)*100)]
#get the column names for each value
col_names = df.columns[60:69]
df3.sort(reverse = True)
#chart size
fig, ax = plt.subplots(figsize = (16,9))
#Horizontal Bar Plot
ax.barh(col_names,df3, color = "green")
#sort column names
ax.invert_yaxis()
#Add numbers to the end of each bar
for i in ax.patches:
    plt.text(i.get_width()+0.2, i.get_y()+0.5,
             str(round((i.get_width()), 2)),
             fontsize = 10, fontweight ='bold',
             color ='grey')
#add title to chart
ax.set_title('How did students find out about computing programs at CCM?')
```

GRAPH NOTE: The numbers pertaining to each x-value is a percentage of students that heard about CCM computing programs from the respective sources.

## Insight #2: 

As we can see in this chart, the best way students found out about computing programs at CCM was info from their counselors and teachers. Other effective ways include the CCM IT Website. One interesting thing is that a small percentage of students found out about computing programs through "CCM Workforce Development" and "NJ Workforce Development". This information tells us that maybe students need to be made more aware of computing programs in these sections, or that we need students to attend more of these events in general.

The following cell is used to create a chart that show the percentages regarding what motivated students to seek a computing degree.

```python
col_names = df.columns[49:59]

motive_list = []
for i in col_names:
    motive_list.append((len(df[df[i] == "Yes"]))/len(df)*100)
motive_list.sort(reverse = True)
#chart size
fig, ax = plt.subplots(figsize = (16,9))
#Horizontal Bar Plot
ax.barh(col_names,motive_list, color = 'Red')
#sort column names
ax.invert_yaxis()
#Add numbers to the end of each bar
for i in ax.patches:
    plt.text(i.get_width()+0.2, i.get_y()+0.5,
             str(round((i.get_width()), 2)),
             fontsize = 10, fontweight ='bold',
             color ='grey')

#Add title
ax.set_title('What motivated students to seek a computing degree at CCM?')
```
